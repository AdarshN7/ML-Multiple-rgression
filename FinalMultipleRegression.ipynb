```python
def ols(x,y):
    n=len(x)
    mean_x=np.mean(x)
    mean_y=np.mean(y)
    numr=np.sum(x*y)-n*mean_x*mean_y  #numerator
    deno=np.sum(x*x)-n*mean_x*mean_x   #denomiantor
    m=numr/deno        #slope
    b=mean_y-m*mean_x   #intercept
    return m,b
```

```python
def tvalue(x,y):
    n=len(x)
    coffs=[]
    coffs=ols(x,y)
    numr=math.sqrt(np.sum((y-coffs[0]*x-coffs[1])**2)/(n-2))
    mean_x=np.mean(x)
    deno=math.sqrt(np.sum((x-mean_x)**2))
    error=round(numr/deno,3)
    if error==0:
        error=numr/deno
    return coffs[0]/error
```

```python
tvalues=[]
for i in range(29):
    x=data.iloc[:,i].values
    t=tvalue(x,y)
    print("X",i+1,"tvalue =",t)
    tvalues.append(t)
```

```python
def calculateR2(y_pred,y):
    numr=np.sum((y-y_pred)**2)
    mean_y=y.mean()
    deno=np.sum((y-mean_y)**2)
    return 1-(numr/deno)
```

```python
calculateR2(y_pred,Y_test)
```

# Gradient Descent

```python
def gradientDescent(theta,x,y):
    y_pred=x.dot(theta)
    lr=0.1
    n=x.shape[1]
    m=x.shape[0]
    diff=[]
    for j in range(n):
        col=x[:,j]
        r=np.sum((y_pred-y)*col)
        r*=lr
        r/=m
        diff.append(r)
    diff1=np.array(diff)
    theta=theta-diff1
    return theta
```

```python
def cost_function(theta,x,y):
    y_pred=x.dot(theta)
    m=x.shape[0]
    sum=np.sum((y_pred-y)**2)
    return sum/(2*m)
```

```python
theta=np.array([0,0,0,0])
```

```python
for i in range(5000):
    theta=gradientDescent(theta,X_train,Y_train)
    print(theta,"cost =",cost_function(theta,X_train,Y_train))
```
